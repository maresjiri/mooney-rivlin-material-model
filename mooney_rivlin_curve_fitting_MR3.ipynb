```python
import numpy as np
import pickle # for loading pickled test data
import pandas as pd # for loading excel data
import matplotlib
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import warnings
from scipy.optimize import differential_evolution
# 
# bounds on parameters are set in generate_Initial_Parameters_MR5() below
def MR3(x, C01, C10, C20):
    #I1uni=(lambda**(2)+2/lambda)
    #I2uni=(2*lambda+2/lambda**(2))
    # x = lambda
    P1 = (2*(2+x)*(1+(1+x)**(2)))/(1+x)**(3)*(C10+C01+2*C20*x**(2)*(2+x)**(2)/(1+x)**(2))
    return P1 

# function for genetic algorithm to minimize (sum of squared error)
# bounds on parameters are set in generate_Initial_Parameters_MR5() below
def sumOfSquaredError_MR3(parameterTuple):
    print("parameterTuple")
    print(parameterTuple)
    print("/parameterTuple")
    warnings.filterwarnings("ignore") # do not print warnings by genetic algorithm
    return np.sum((yData - MR3(xData, *parameterTuple)) ** 2)

def generate_Initial_Parameters_MR3():
    parameterBounds = []
    parameterBounds.append([-100000.0, 100000.0]) # parameter bounds for C_01
    parameterBounds.append([-100000.0, 100000.0]) # parameter bounds for C_10
    parameterBounds.append([-100000.0, 100000.0]) # parameter bounds for C_20
    result = differential_evolution(sumOfSquaredError_MR3, parameterBounds, seed=3)
    return result.x

# load the pickled test data from original Raman spectroscopy
#data = pickle.load(open('data.pkl', 'rb'))
#xData = data[0]
#yData = data[1]

#data = pd.read_excel('experimental_data1.xlsx', sheet_name='ExperimentData')
data = pd.read_excel('rubber3mm.xlsx', sheet_name='Sheet1')

xData = data['lmbd'].values
yData = data['P'].values

print(xData)
print(yData)

# generate initial parameter values
initialParameters = generate_Initial_Parameters_MR3()

print("initialParameters")
print(initialParameters)
print("/initialParameters")


# curve fit the test data
fittedParameters, pcov = curve_fit(MR3, xData, yData, initialParameters)

print("pcov")
print(pcov)
print("/pcov")

print("fittedParameters")
print(fittedParameters)
print("/fittedParameters")


# create values for display of fitted peak function
C01, C10, C20 = fittedParameters
y_fit = MR3(xData, C01, C10, C20)

print("y_fit")
print(y_fit)
print("/y_fit")

plt.plot(xData, yData) # plot the raw data
plt.plot(xData, y_fit) # plot the equation using the fitted parameters
plt.show()

print(fittedParameters)
print("end")
```
